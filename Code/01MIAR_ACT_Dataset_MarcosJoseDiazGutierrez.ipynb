# 0. Introducción al dataset.

## 0.1. Descripción General.

 Este dataset ofrece una visión detallada de los estudiantes matriculados en distintas carreras de grado en una institución de educación superior. Incluye información demográfica, factores socioeconómicos y datos sobre el rendimiento académico, que pueden ayudar a analizar posibles factores predictivos de abandono estudiantil y éxito académico. El conjunto está compuesto por múltiples bases de datos independientes que recopilan información relevante en el momento de la inscripción, como el modo de aplicación, estado civil, carrera elegida, entre otros. Además, estos datos permiten estimar el rendimiento de los estudiantes al final de cada semestre evaluando las unidades curriculares acreditadas, matriculadas, evaluadas y aprobadas, así como sus calificaciones. También se incluyen indicadores económicos como la tasa de desempleo, inflación y el PIB de la región, los cuales permiten entender mejor la influencia de factores económicos en el abandono o éxito académico. Este análisis puede aportar una perspectiva valiosa sobre lo que impulsa a los estudiantes a continuar con sus estudios o abandonarlos en una gran variedad de disciplinas, como agronomía, diseño, educación, enfermería, periodismo, administración, trabajo social o tecnología.

 El dataset se puede encontrar en Kaggle: https://www.kaggle.com/code/jeevabharathis/student-dropout-analysis-for-school-education/input

## 1.3. Corregir campos mal formateados: *Curricular units 1st sem (grade)*, *Curricular units 2nd sem (grade)*

Estos campos vienen mal formateados. Toman valores numericos separados por '.', que no tienen sentido aparente. Después de estudiar estos valores y de investigar en Internet esta es la explicación más probable: en realidad estos campos deberían ser números del 0 al 20, que por algún motivo, se han formateado mal con los decimales. Como contraejemplo, tenemos dos valores superiores al 20 que refutarían esta hipotesis (a no ser que consideremos que el valor 114 era en realidad 11.4).

```python
test_values = ['13.428.571.428.571.400', '14.0', '0.0', '20.0', '114.375']

def clean_and_convert_grade(value):
    try:
        value_str = str(value)

        # Indice del primer punto
        first_point_index = value_str.find('.')
        
        # Hay veces que no hay ningun punto...
        if first_point_index == -1:
            cleaned_value = value_str
        else:
            # Cogemos la primera parte, y añadimos el resto de partes quitando el punto
            cleaned_value = value_str[:first_point_index + 1] + value_str[first_point_index + 1:].replace('.', '')
        
        float_value = round(float(cleaned_value), 2)
        
        # Aplicamos lo que deciamos antes de los valores mayores de 20
        if float_value > 20:
            float_value = round(float_value / 10, 2)
        
        return float_value
    
    except:

        return float('nan')


df['Curricular units 1st sem (grade) new'] = df['Curricular units 1st sem (grade)'].apply(clean_and_convert_grade)
df[['Curricular units 1st sem (grade) new', 'Curricular units 1st sem (grade)']]
```

Viendo que funciona (después de varios intentos), aplicamos la funcion a los campos correspondientes.

```python
df.drop(columns=['Curricular units 1st sem (grade) new'])
df['Curricular units 1st sem (grade)'] = df['Curricular units 1st sem (grade)'].apply(clean_and_convert_grade)
df['Curricular units 2nd sem (grade)'] = df['Curricular units 1st sem (grade)'].apply(clean_and_convert_grade)

print(df[['Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)']].isna().sum())
df[['Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)']]
```

# 2. Visualización de Datos.

Para tener una visión integral del dataset, vamos a pintar las distribuciones de cada campo (eligiendo adecuadamente en función del tipo de dato) distinguiendo además entre los casos *Dropout* y los *No-Dropout*

```python
columns_to_plot = df.columns.drop("Target")

plot_types = {col: "categorical" if str(df[col].dtype) == "category" else "numerical" for col in columns_to_plot}

for col in columns_to_plot:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
    fig.suptitle(f"Distribución de {col} por Target (Proporción: No-Dropout vs Dropout)")
    
    if plot_types[col] == "categorical":
        no_dropout_data = df[df["Target"] == "No-Dropout"][col].value_counts(normalize=True)
        sns.barplot(x=no_dropout_data.index, y=no_dropout_data.values, ax=axes[0])
    else:
        sns.histplot(data=df[df["Target"] == "No-Dropout"], x=col, kde=True, stat="density", ax=axes[0])
    
    axes[0].set_title(f"No-Dropout")
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("Proportion")
    
    if plot_types[col] == "categorical":
        dropout_data = df[df["Target"] == "Dropout"][col].value_counts(normalize=True)
        sns.barplot(x=dropout_data.index, y=dropout_data.values, ax=axes[1])
    else:
        sns.histplot(data=df[df["Target"] == "Dropout"], x=col, kde=True, stat="density", ax=axes[1])
    
    axes[1].set_title("Dropout")
    axes[1].set_xlabel(col)
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

## 2.1. Comentarios de las visualizaciones más relevantes.

- **Marital Status**: vemos que la proporción de la clase 1 es menor en Dropouts que en No-Dropouts. Esto sugiere que la clase 1 sea soltero (en la universidad poca gente se ha casado aún) y la clase 2 sea casado. Y, que los casados dejen la universidad en mayor proporción de lo que lo hacen los no casados. 

- **Application mode**: aquí llama la atención la clase 12. La proporción de dropouts que han aplicado a través de la vía 12 es mucho superior a la proporción de no-dropouts que han aplicado a través de la vía 12. ¿Qué es esta vía? No lo podemos saber; pero, podemos conjeturar que la vía 12 es un tipo de aplicación reservada para personas con algún tipo de necesidad especial (social, económica, física, etc). 

- **Displaced**: aquí vemos distribuciones espejo. La clase 0 es proporcionalmente menos frecuente en no-dropout que en dropout; análogamente con la clase 1 ya que sólo hay dos clases en este campo. Es decir, si elegimos un estudiante al azar y resulta que es dropout, es más probable que sea Displaced = 0 que Displaced = 1 (y viceversa). Aquí parece que tiene sentido conjeturar de nuevo que el valor 0 significa que el alumno no ha sido desplazado (menor proporción de dropout) y que 1 significa que sí ha sido desplazado (mayor proporción de dropout).

Es fundamental recordar que estas observaciones son correlaciones y no necesariamente causas. La mayor proporción de abandono en ciertos grupos (como estudiantes casados o desplazados) podría estar relacionada con factores que aún desconocemos, y no implica que ser parte de esos grupos sea la causa del abandono. Es importante investigar más a fondo antes de sacar conclusiones sobre la causalidad. **¡Correlación no implica causalidad!**


# 3. Tratamiento de Datos

## 3.1. Campos categóricos.

```python
# Hacemos dummies, ya que no tenemos tantos campos categoricos, y estos campos no tienen tantos valores unicos
print(f'Antes de dummies: df.shape: {df.shape}')
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
print(f'Despues de dummies: df.shape: {df.shape}')
```

## 3.2. Campos numéricos

```python
def scale_and_save(df, numerical_columns, folder='min_max'):
    os.makedirs('min_max', exist_ok=True)
    df_scaled = df.copy()
    
    for col in numerical_columns:
        scaler = MinMaxScaler()
        df_scaled[col] = scaler.fit_transform(df[[col]])
        
        scaler_filename = os.path.join(folder, f"{col}_scaler.pkl")
        with open(scaler_filename, 'wb') as f:
            pickle.dump(scaler, f)
    
    return df_scaled

# Esta funcion la usaremos con el X_train para no hacer fuga de datos
```

# 4. train_test_split

```python
X = df.drop(columns=['Target_No-Dropout'])
y = df['Target_No-Dropout']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

X_train = scale_and_save(df=X_train, numerical_columns=numerical_columns)

# Usamos los objetos del train para el test
X_test = scale_and_save(df=X_test, numerical_columns=numerical_columns)
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "SVM": SVC()
}

results = []

for model_name, model in models.items():
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    results.append(pd.DataFrame({"Model": model_name, "Accuracy": cv_scores}))

results_df = pd.concat(results, ignore_index=True)

plt.figure(figsize=(12, 6))
sns.boxplot(x="Model", y="Accuracy", data=results_df)
plt.title("validacion cruzada para diferentes modelos")
plt.ylabel("Accuracy")
plt.xlabel("Modelo")
plt.show()
```

# 5. Conclusiones.

Como podemos ver en el gráfico anterior, el modelo más robusto (con menos variabilidad de accuracy y mejor accuracy) es el de regresión logística. Hay veces, que por mucho que nos empeñemos en complicar las cosas, las cosas más sencillas son las que mejor funcionan. Para ejemplo: este caso. 

## 5.1. Variables más importantes.

```python
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

coef = logreg.coef_.flatten()  
features = X_train.columns     

importance_df = pd.DataFrame({
    'Feature': features,
    'Coefficient': coef
}).sort_values(by='Coefficient', key=np.abs, ascending=False)

plt.figure(figsize=(15, 60))
plt.barh(importance_df['Feature'], importance_df['Coefficient'], color="skyblue")
plt.xlabel("Coefficient Value")
plt.ylabel("Feature")
plt.title("Feature Importance in Logistic Regression Model")
plt.gca().invert_yaxis()
plt.show()
```
